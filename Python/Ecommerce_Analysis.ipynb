```python
# The next step is indeed Data Cleaning, but in this order:
# Never modify the original data.
```

```python
customers_clean = customers.copy()
orders_clean = orders.copy()
products_clean = products.copy()
order_items_clean = order_items.copy()
payments_clean = payments.copy()
reviews_clean = reviews.copy()
sellers_clean = sellers.copy()
geolocation_clean= geolocation.copy()
```

```python
orders_clean.select_dtypes(include='object').columns
```

```python
date_col = [
    'order_purchase_timestamp',
    'order_approved_at', 
    'order_delivered_carrier_date',
    'order_delivered_customer_date', 
    'order_estimated_delivery_date'
    ]

for i in date_col:
    orders_clean[i] = pd.to_datetime(orders_clean[i])
```

```python
orders_clean[date_col].dtypes
```

```python
orders_clean['order_year'] = orders_clean['order_purchase_timestamp'].dt.year
orders_clean['order_month'] = orders_clean['order_purchase_timestamp'].dt.month
orders_clean['order_month_name'] = orders_clean['order_purchase_timestamp'].dt.month_name()
orders_clean['order_day_name'] = orders_clean['order_purchase_timestamp'].dt.day_name()
orders_clean['order_quarter'] = orders_clean['order_purchase_timestamp'].dt.quarter
```
